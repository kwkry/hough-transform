# Hough Transform

**Importing Modules**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Using Rubix Cube

###**Import Image**

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
filepath = '/content/drive/MyDrive/Digital Signal Processing/Group 8 - Hough Transform/Images/Rubix Cube.jpg'
rubix_img = cv2.imread(filepath)
```

###**Transform Image to Grayscale**

```python
figsize = (5,5)
rubix_gray = cv2.cvtColor(rubix_img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=figsize)
plt.imshow(rubix_gray, cmap="gray", vmin=0, vmax=255)
plt.show()
```

###**Apply Gaussian Blur**

```python
rubix_blurred = cv2.GaussianBlur(rubix_gray, (5, 5), 0)

plt.figure(figsize=figsize)
plt.imshow(rubix_blurred, cmap="gray", vmin=0, vmax=255)
plt.show()
```

###**Apply Canny Edge Detection**

```python
rubix_edges = cv2.Canny(rubix_blurred, 50, 25)

plt.figure(figsize=figsize)
plt.imshow(rubix_edges)
plt.show()
```

###**Apply Hough Transform**

```python
rubix_lines = cv2.HoughLines(rubix_edges, 1, np.pi / 180, threshold=200)
```

**Draw Hough Lines**

```python
if rubix_lines is not None:
    for rho, theta in rubix_lines[:, 0]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1500 * (-b))
        y1 = int(y0 + 1500 * (a))
        x2 = int(x0 - 1500 * (-b))
        y2 = int(y0 - 1500 * (a))
        cv2.line(rubix_img, (x1, y1), (x2, y2), (255, 0, 0), 3)
```

**Display Results**

```python
plt.figure(figsize=figsize)
plt.imshow(rubix_img)
plt.show()
```

## Using Road Lines

```python
figsize = (5,5)
roadLines_gray = cv2.cvtColor(roadLines_img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=figsize)
plt.imshow(roadLines_gray, cmap="gray", vmin=0, vmax=255)
plt.show()
```

```python
if roadLines_lines is not None:
    for rho, theta in roadLines_lines[:, 0]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))
        cv2.line(roadLines_img, (x1, y1), (x2, y2), (255, 0, 0), 2)
```
